Define <tt>BiasPmf</tt>. Note that this is the code from your book.

```python
def BiasPmf(pmf, label=''):
    """Returns the Pmf with oversampling proportional to value.

    If pmf is the distribution of true values, the result is the
    distribution that would be seen if values are oversampled in
    proportion to their values; for example, if you ask students
    how big their classes are, large classes are oversampled in
    proportion to their size.

    Args:
      pmf: Pmf object.
      label: string label for the new Pmf.

     Returns:
       Pmf object
    """
    new_pmf = pmf.Copy(label=label)
    new_pmf.label = label
    for x, p in pmf.Items():
        new_pmf.Mult(x, x)
        
    new_pmf.Normalize()
    return new_pmf
```

Make the biased Pmf chart of children in the household, as observed if you surveyed the children instead of the respondents.

```python
bias_pmf = BiasPmf(und18_pmf, label='observed')
thinkplot.Config(xlabel='Number of Children', ylabel='PMF', title="Observed Number of children per household")
thinkplot.Pmf(bias_pmf)
thinkplot.Show()
```

The actual and the observed are different because of the difference in the number of observers. For example

In the case where there is 0 children:
* Actual shows that about 47% of families have no children.
* Observed shows 0%, because there are no children to make this observation

Looking at the case where there is 2 children in the family is more useful to explain the phenomenon:
* Actual shows that about 20% of families have two children. For each family with two children, you get a report of two children.
* However, when you ask a family that has two children how many children there are, you will get two reports of two children.

This is annoying to explain. But basicaly when you are trying to find a biased/observed value, you are get N reports that say there are N objects. Where actual would just report that there are N objects.
